```python
import random
import matplotlib.pyplot as plt

def linear_search(arr, key):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == key:
            return comparisons
    return comparisons

input_sizes = [100, 500, 1000, 5000, 10000]
results = {'Best': [], 'Average': [], 'Worst': []}

for size in input_sizes:
    arr = list(range(size))
    results['Best'].append(linear_search(arr, arr[0]))
    results['Average'].append(linear_search(arr, arr[size // 2]))
    results['Worst'].append(linear_search(arr, -1))

plt.plot(input_sizes, results['Best'], marker='o', label='Best Case')
plt.plot(input_sizes, results['Average'], marker='o', label='Average Case')
plt.plot(input_sizes, results['Worst'], marker='o', label='Worst Case')

plt.xlabel('Input Size (n)')
plt.ylabel('Number of Comparisons')
plt.title('Linear Search Comparisons vs Input Size')
plt.legend()
plt.grid(True)
plt.show()
```
